Preprocessing is the process of doing a pre-analysis of data, in order to transform them into a standard and normalized format.

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandas_dataset.csv")
df.head(5)
```

```python
df["Date"]= df["Date"].str.replace("'"," ")
df
```

```python
#return a dataset with no empty cells
#df.dropna() will remove empty cells but wll not change the dataframe
# if we want dataframe to be changed,use replace = true

new_df = df.dropna()
new_df
```

```python
#replace the nan with the value 130
new_df2=df.fillna(130)
new_df2
```

```python
df["Calories"].fillna(140,inplace=True)
df
```

```python
#fill null values witn mean,median or mode

import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandas_dataset.csv")
#find the mean value of calories
x=df['Calories'].mean()
#replace the nan value with the mean value calculated
df["Calories"]=df["Calories"].fillna(x)
df
```
